```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""
[[  12    4    2    2    7    6]
 [ 140 8022  281  211 1679 1699]
 [   7   16  171   11   22   10]
 [  28   25   41  585   45   23]
 [ 210 1153  289  383 5293 1159]
 [  20  426   20   39  232  462]]
"""

gyvunams = [12, 4, 2, 2, 7, 6]
moterims = [140, 8022, 281, 211, 1679, 1699]
namams = [7, 16, 171, 11, 22, 10]
pramogos = [28, 25, 41, 585, 45, 23]
vaikams = [210, 1153, 289, 383, 5293, 1159]
vyrams = [20, 426, 20, 39, 232, 462]


cm = np.array([gyvunams, moterims, namams, pramogos, vaikams, vyrams])
labels = ["Gyvnams", "Moterims", "Namams", "Pramogos", "Vaikams", "Vyrams"]

cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)
sns.heatmap(
    cm_norm,
    annot=True,
    cmap=sns.cubehelix_palette(as_cmap=True),
    xticklabels=labels,
    yticklabels=labels,
)
plt.xlabel("Spjama reikm")
plt.ylabel("Tikra reikkm")
plt.title("Normalizuota klasifikavimo matrica")
plt.savefig("logloss_conf_matrix.svg", format="svg", bbox_inches="tight")
plt.show()
```
